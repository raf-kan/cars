```python
import pandas as pd
import numpy as np
```

# Load data

```python
filename = 'data/imports-85.data'
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names=headers)
```

```python
df.head(5)
```

# Cleaning the Data

### Replace '?' with NaN

```python
df.replace("?", np.nan, inplace=True)
df.head(5)
```

### Is there missing data?

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
table_missing_data = pd.DataFrame()
```

```python
for col in missing_data.columns.values.tolist():
    table_missing_data[col] = missing_data[col].value_counts()
```

```python
table_missing_data
```

```python
missing_data.describe()
```

### Replace the missing data

```python
# Replacing that data with it's average

# To avoid multiple lines of code, a simple function that calls in the dataframe and the column to change will do fine
def replace_with_avg(df, col_name):
    avg = df[col_name].astype('float').mean(axis=0)
    df[col_name].replace(np.nan, avg, inplace=True)
    return df
```

```python
df = replace_with_avg(df, 'peak-rpm')
df = replace_with_avg(df, 'normalized-losses')
df = replace_with_avg(df, 'bore')
df = replace_with_avg(df, 'stroke')
```

```python
df.head(5)
```

```python
# Addressing data containing strongs of values
# Replace by frequency in data

df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
```

```python
# Drop rows that have missing key data
df.dropna(subset=['price'], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
df[['bore','stroke']] = df[['bore','stroke']].astype('float')
df[['normalized-losses']] = df[['normalized-losses']].astype('int')
df[['price']] = df[['price']].astype('float')
df[['peak-rpm']] = df[['peak-rpm']].astype('float')
```

```python
df.dtypes
```

```python
df.to_csv('data/cars.csv', index=False)
```
